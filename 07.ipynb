```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/ADMIN/Documents/img/01/5.jpg'
# Đọc ảnh và hiển thị
im = Image.open(my_path)
# display(im)
plt.imshow(im)
```

```python
# Vẽ đồ thị cơ bản
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]
plt.plot(x, y, color="green") # màu sắc
plt.title('Ví dụ về biểu đồ') # Title
# Tắt hệ trục tọa độ
plt.axis('off')  # Tắt hệ trục tọa độ
plt.show()
```

```python
# Vẽ trên hình ảnh
# ĐỌc ảnh
im = Image.open(my_path)
x = [100, 100, 400, 400]
y = [200, 300, 200, 300]

plt.imshow(im) # Hiển thị hình ảnh
plt.plot(x,y, 'bo-') 
plt.show()
```
